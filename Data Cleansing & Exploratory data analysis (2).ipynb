```python
from sklearn.linear_model import LinearRegression
import math
df.reset_index(inplace = True)
```

```python
# Using dat of prodcuts with a Rating as training set for  linear regression model based on price vs Rating relation
# Using non zero values of rating as training data set, and price data for Ratings value of 0 as Test data,  this is more than the data split of 90:10
x_train = df[df['Rating(5)'] != 0]['Price'].to_frame().reset_index()
y_train = df[df['Rating(5)'] != 0]['Rating(5)']
x_test = df[df['Rating(5)'] == 0]['Price'].to_frame().reset_index()
model = LinearRegression()
model.fit(x_train,y_train)
Prediction = model.predict(x_test)
```

```python
d = {'Product' : df[df['Rating(5)'] == 0]['Index'],'Price' : df[df['Rating(5)'] == 0]['Price'], 'Predicted_Ratings' : Prediction}
Prediction_df = pd.DataFrame(data = d)
Prediction_df['Predicted_Ratings'] = round(Prediction_df['Predicted_Ratings'],1)
# Most rating values are same or just repeating values, clearly indicating Price of products and Rating doesnt have a linear relationship
```

```python
#Using multiple variables for Linear regression
# We can use number of Number_of_ratings as the additional variable
# If we look at the number of unique datas before filling non type objects with 0, saved before
# Detailed Description                278
# Price                               163
# Rating(5)                            30
# Number_of_ratings                   139
# Bought_in_past_month (exceeding)     14
# Link                                278

# Hence number of entries for using this linear regression model would further decrease
# Using non zero values of Number_of_ratings as training data set, and price data for Ratings value of 0 as Test data, this is more than the data split of 90:10
#  Here rating is taken as the feature whose values we predict based on the regresion model, as Rating implies the probablity that we buy the product

x1_train = df[df['Number_of_ratings'] != 0][['Price','Number_of_ratings']]
y1_train = df[df['Number_of_ratings'] != 0]['Rating(5)']
x1_test = df[df['Number_of_ratings'] == 0][['Price','Number_of_ratings']]

model = LinearRegression()
model.fit(x1_train,y1_train)

Prediction_1 = model.predict(x1_test)
# When we look at Predection_1 all ratings are 3.8 or 3.9 if rounded off, this is not appropriate
# We have very less number of datapoints, thats probably giving us a inaccurate regression model
# Number of ratings and rating are not independant variables hence they might not give an accurate prediction if multiple variables are used for regression model
# We could extract more data from Amazon like dimensions of the pump, power etc, which could give more specificity to data and we could use Kmeans clustering to find ratings of data more accurately
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
km_df = df[['Price','Rating(5)','Number_of_ratings']]
scaler = MinMaxScaler()
```

```python
scaler = MinMaxScaler()

scaler.fit(km_df[['Price']])
km_df['Price'] = scaler.transform(km_df[['Price']])

scaler.fit(km_df[['Rating(5)']])
km_df['Rating(5)'] = scaler.transform(km_df[['Rating(5)']])

scaler.fit(km_df[['Number_of_ratings']])
km_df['Number_of_ratings'] = scaler.transform(km_df[['Number_of_ratings']])
```

```python
kmeans = KMeans(n_clusters=4)
#kmeans
```

```python
y_predicted = kmeans.fit_predict(km_df)
y_predicted
```

```python
km_df['Cluster'] = y_predicted
```

```python
sse =[]
k_rng = range(1,30)
for k in k_rng:
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(km_df)
    sse.append(kmeans.inertia_)
```
