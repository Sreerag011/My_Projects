```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\user\OneDrive\Documents\Car_Washer_Data.csv')
```

```python
#Data cleansing
#dropped the Index column carried over from Extracted csv file
df = df.drop(columns = 'Unnamed: 0')

#setting no of rows displayed as 1000 so we could look at all the rows
pd.set_option('display.max.rows',1000)
#pd.set_option('display.max.columns',30)
df.info()
```

```python
#to drop duplicate rows
df = df.drop_duplicates()
df.info()
#No duplicates found,as there is no change in number of rows
```

```python
df
```

```python
#Rating is out of 5 so only the first number is needed, which is rounded to one decimal place already
df['Rating'] = df['Rating'].str[0:3]

#Renaming Column Rating as Rating (5) & title as Title and Bought_in_past_month as Bought_in_past_month (exceeding)
df = df.rename(columns = {'Rating' : 'Rating(5)', 'title' : 'Detailed Description' ,'Bought_in_past_month' : 'Bought_in_past_month (exceeding)'})
```

```python
#Bought_in_past_month can be stripped to to get just numbers, as the title already suggested what the number means
df['Bought_in_past_month (exceeding)'] = df['Bought_in_past_month (exceeding)'].str.replace('+ bought in past month','')
#again K needs to be replaced with 1000
df['Bought_in_past_month (exceeding)'] = df['Bought_in_past_month (exceeding)'].str.replace('K','000')
```

```python
#Abbreviation of Title for ease plotting later on, abbreviated to 7 letters and used as Index
#df['Index'] = df['Detailed Description'].str[0:6]
#df.set_index('Index',inplace = True)
#changed later to avoid duplication of Index
```

```python
#df.loc['AGARO ']
# Checking for this showed 3 issues with the list,
#   1.Name abbreviation has to be done for more charecters to avoid duplication of Index
#   2.Because of Links being different some duplicate values were not removed by drop_duplicates function.
#     So need to remove duplicates having values same except link, as link varies for same product in just some parameters.
#   3.Need to strip of or replace White spaces and special charecters with Under scores,for Abbrevations to have unique Index values

df=df.drop_duplicates(subset =['Detailed Description','Price','Rating(5)','Number_of_ratings','Bought_in_past_month (exceeding)'] )
df.info()
# Number of Enties in the list redcued to almost Half, as duplicates values were removed.

#Taking 14 charecters from Details description to use as Index for ease of plotting
df['Index'] = df['Detailed Description'].str[0:14]

#Stripping it of white spaces
df['Index']= df['Index'].str.strip()

#Replacing special charecters with under score, except '|áÂ'
df['Index'] = df['Index'].str.replace('[^a-zA-Z0-9|áÂ]','_',regex = 'True')

#striping under_score/hyphen/pipe(|) from right side
df['Index'] = df['Index'].str.rstrip('_|-')
df['Index'].str[13:14]

#On displaying the last charecter of the Index column,there are no special charecters at end but white spaces, hence using rstrip()
df['Index'] = df['Index'].str.rstrip()
#It could also mean there is no 14th Charecter,confirmed by : print(df['Index'])

#finallly setting Index
df.set_index('Index',inplace = True)


# Most rows in Bought_in_past_month (exceeding) is blank.
# Not dropping the column as we can use the available 140 Prodcuts with this data as training set to predict,
# the possibility of other products being bought in next month, aslo to cross-check the best available Prodcut, based on our data analysis.
```

```python
#converting numerical columns to appropriate datatypes

#cannot convert to integer as some value has a period/comma in it '.,', hence using

df['Price'] = df['Price'].str.replace('[.,]','',regex = True)
df['Price'] = df['Price'].astype('int64')
```

```python
#formating Rating column to avoid commas and white spaces
df['Rating(5)'] = df['Rating(5)'].replace(',','',regex = True)
df['Rating(5)'] = df['Rating(5)'].str.strip()
df['Rating(5)'] = df['Rating(5)'].astype('float')
df.info()
```

```python
df['Number_of_ratings'] = df['Number_of_ratings'].str.replace('[.,]','',regex = True)

#cannot convert float Nan to integer
df['Number_of_ratings'] = df['Number_of_ratings'].astype('float')
# Now we can convert float 0.0 to integer 0
df['Number_of_ratings']= df['Number_of_ratings'].fillna('0')
```

```python
df['Number_of_ratings'] = df['Number_of_ratings'].astype('int64')
df.info()
```

```python
#df['Bought_in_past_month (exceeding)'] = df['Bought_in_past_month (exceeding)'].astype('int64')

#cannot convert float Nan to integer
df['Bought_in_past_month (exceeding)'] = df['Bought_in_past_month (exceeding)'].astype('float')
# Now we can convert float 0.0 to integer 0
df['Bought_in_past_month (exceeding)']= df['Bought_in_past_month (exceeding)'].fillna('0')
```

```python
df['Bought_in_past_month (exceeding)'] = df['Bought_in_past_month (exceeding)'].astype('int64')
df.info()
```
